```python
with open("data.csv") as f:
    data = f.read().splitlines()
data = list(map(lambda x: list(map(float, x.split(","))), data))
```

```python
data
```

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.figure(0)
plt.subplot(141)
plt.scatter(list(map(lambda x: x[0], data)), list(map(lambda x: x[4], data)), label="appreciation")
plt.subplot(142)
plt.scatter(list(map(lambda x: x[1], data)), list(map(lambda x: x[4], data)), label="house annual cost")
plt.subplot(143)
plt.scatter(list(map(lambda x: x[2], data)), list(map(lambda x: x[4], data)), label="market return")
plt.subplot(144)
plt.scatter(list(map(lambda x: x[3], data)), list(map(lambda x: x[4], data)), label="interest")
plt.legend()
plt.show()
```

```python
# appr: N(5,1)
# cost: N(2,1)
# return: N(7,4)
# interest: N(3.75, 1)
plt.figure(1)
plt.suptitle("sensitivity of (wealth buy - wealth rent) w.r.t. various inputs")
plt.subplot(141).title.set_text("appreciation")
mdata = list(filter(lambda x: 1 <= x[1] <= 3 and 5<= x[2] <= 9 and 3 <= x[3] <= 4.5, data))
plt.scatter(list(map(lambda x: x[0], mdata)), list(map(lambda x: x[4], mdata)))

plt.subplot(142).title.set_text("house annual cost")
mdata = list(filter(lambda x: 4 <= x[0] <= 6 and 5<= x[2] <= 9 and 3 <= x[3] <= 4.5, data))
plt.scatter(list(map(lambda x: x[1], mdata)), list(map(lambda x: x[4], mdata)))

plt.subplot(143).title.set_text("market return")
mdata = list(filter(lambda x: 4 <= x[0] <= 6 and 1 <= x[1] <= 3 and 3 <= x[3] <= 4.5, data))
plt.scatter(list(map(lambda x: x[2], mdata)), list(map(lambda x: x[4], mdata)))

plt.subplot(144).title.set_text("interest")
mdata = list(filter(lambda x: 4 <= x[0] <= 6 and 1 <= x[1] <= 3 and 5<= x[2] <= 9, data))
plt.scatter(list(map(lambda x: x[3], mdata)), list(map(lambda x: x[4], mdata)))
plt.show()
```
